```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
cv2.__version__
```

```python
img1 = cv2.imread('/Users/thersajeanbaptiste/REU--PYTHON/REU-PYTHON/forest.jpg')
img2 = cv2.imread('/Users/thersajeanbaptiste/REU--PYTHON/REU-PYTHON/people-walking.jpg')
print('showing size image1' , img1.shape)
print('showing size image2',img2.shape)
img1_reshaped = cv2.resize(img1,(300,300), interpolation = cv2. INTER_AREA)
img2_reshaped = cv2.resize(img2,(300,300), interpolation = cv2. INTER_AREA)
print('showing size images2 reshaped' ,img2_reshaped.shape)
```

```python
#showing images
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.imshow(img1_reshaped)
plt.subplot(222)
plt.imshow(img2_reshaped)
```
